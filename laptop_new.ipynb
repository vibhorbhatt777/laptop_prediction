```python
new_data = new_data.drop(['Os', 'TypeName', 'Company'], axis=1)
```

```python
print(new_data['Cpu_brand'].value_counts())
```

```python
cpu_dict = {
    'Other Intel Processor': 0,
    'Intel Core i3': 1,
    'AMD Processor': 2,
    'Intel Core i5': 3,
    'Intel Core i7': 4
}
gpu_dict = {
    'AMD': 0, 'Intel': 1, 'Nvidia': 2
}
```

```python
new_data['Cpu_brand'] = new_data['Cpu_brand'].map(cpu_dict)
new_data['Gpu_brand'] = new_data['Gpu_brand'].map(gpu_dict)
```

```python
new_data
```

```python
num_data
```

```python
num_data['Weight'] = num_data['Weight'].round().astype('int')
```

```python
def categorize_Ppi(ppi):
    if 90 <= ppi < 143: return 0
    elif 143 <= ppi < 196: return 1
    elif 196 <= ppi < 249: return 2
    elif 249 <= ppi < 302: return 3
    else: return 4
```

```python
num_data['Ppi'] = num_data['Ppi'].apply(categorize_Ppi)
```

```python
complete_df = pd.concat([new_data, num_data], axis=1)
complete_df.head()
```

```python
complete_df['SSD'].value_counts()
```

```python
complete_df = complete_df[~complete_df['HDD'].isin([32, 128])]
complete_df = complete_df[~complete_df['SSD'].isin([180, 16,1024, 768, 240, 8, 64, 32])]
complete_df
```

```python
complete_df.corr()['Price']
```

```python
complete_df = complete_df.drop('HDD', axis=1)
complete_df
```

```python
x = complete_df.drop('Price', axis=1)
y = complete_df[['Price']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=91)
rdf = RandomForestRegressor()
```

```python
rdf.fit(x_train, y_train)
```

```python
rdf.score(x_train, y_train)
```

```python
rdf.score(x_test, y_test)
```

```python
pred = rdf.predict(x_test)
compare = y_test
compare['Prediction'] = pred
compare
```
